```python
# Code for data preprocessing, loading a dataset, and visualizations
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Sample Dataset: Iris Dataset (can be loaded from sklearn or a CSV file)
from sklearn.datasets import load_iris

data = load_iris()
iris = pd.DataFrame(data.data, columns=data.feature_names)
iris['species'] = data.target
```

```python
# Handling missing value
iris.loc[5:10, 'species'] = np.nan
iris['species'] = iris['species'].fillna(iris['species'].mode()[0])
```

```python
# Encoding categorical data
iris['species'] = iris['species'].astype(int)
```

```python
# Feature scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
iris_scaled = scaler.fit_transform(iris.iloc[:, :-1])
```

```python
# Visualization: Histogram
plt.hist(iris[data.feature_names[0]], bins=10, color='blue', alpha=0.7)
plt.title('Feature Distribution')
plt.show()
```

```python
# Scatter plot
sns.scatterplot(x=data.feature_names[0], y=data.feature_names[1], hue='species', data=iris)
plt.title('Scatter Plot')
plt.show()
```

```python
# Correlation heatmap
sns.heatmap(iris.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
